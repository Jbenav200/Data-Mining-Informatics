```python
x_train, x_test, y_train, y_test = train_test_split(features, classes, test_size= .33, random_state=4450)
print(x_train.shape, x_test.shape)
```

```python
decisionTree = DecisionTreeClassifier(criterion='entropy')
print(decisionTree)
```

```python
dtc_model = decisionTree.fit(x_train, y_train)
```

```python
importance = dtc_model.feature_importances_
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))
# Barchat for feature importance
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```

```python
prediction = dtc_model.predict(x_test)
```

```python
y_true = le.inverse_transform(y_test["Loan_Status"])
y_pred = le.inverse_transform(prediction)
```

```python
cm = confusion_matrix(y_true, y_pred)
labels = ['N', 'Y']
ConfusionMatrixDisplay(cm, display_labels=labels).plot()
```

```python
print(classification_report(y_true, y_pred))
```

The result of removing the above mentioned attributes from the list of features, meant that fewer attributes were used to produce the final tree. This often results in a better classification model as it simplifies the process of decision making for the algorithm. These particular attributes were removed after calculating the feature importance for the given set of features. After removing each attribute, the next lowest scoring attribute was selected for removal and the decision tree built and a confusion matrix applied to test the accuracy of the decision tree based on the feature set selected.

With this approach, multiple decision tree stuctures were built and their accuracy checked, which did take more time. However, the end result was as optimal a decision tree I could create based on the attributes within the dataset and the features I decided to use.

The first decision tree utilised all of the attributes, as a result there were a lot of attributes contributing to the decision tree that didn't necessarily affect whether or not someone is likely to repay a loan. This meant that the decision tree was overfitted, and wouldn't necessarily correctly classify new data that is not in the testing and training sets.

The latter decission tree utilised fewer features, meaning fewer attributes were being used to decide on whether a loan was safe to give to an applicant or not. There were attributes which were removed from the list of features which might have been beneficial to the decision tree to keep in, like CoapplicantIncome for example. In most cases if there is a co-applicant, chances are their income will affect the amount you will be able to take out as a loan and the security the lender feels in giving out the loan. As a result, this decision tree could suffer from underfitting, meaning that it might not correctly classify new data that is not in the testing and training sets.
